```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing the Boston House Price Dataset

```python
house_price_dataset = pd.read_csv("/Boston.csv")
house_price_dataset.head()
```

```python
#Removing the "Unnamed: 0" column from the data
house_price_dataset = house_price_dataset.drop(['Unnamed: 0'], axis=1)
```

```python
#Checking the rows and columns of the data
house_price_dataset.shape
```

```python
house_price_dataset.head()
```

```python
# Checking for missing values
house_price_dataset.isnull().sum()
```

```python
# Statistical measures of the dataset
house_price_dataset.describe()
```

Understanding the Coorelation between various features in the dataset

```python
correlation = house_price_dataset.corr()
```

```python
# Construting a Heatmap to understand the correlation

fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues');
```

Spliting the data and the Target

```python
x = house_price_dataset.iloc[:,:-1]
y = house_price_dataset["price"]
```

Spiliting the Training data and Test data

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

Model Training


XGBOOST Regression

```python
# Loading the model
model = XGBRegressor()
```

```python
#training the model with x_train and y_train
model.fit(x_train, y_train)
```

Evaluation

```python
#Checking the prediction score of the model
model.score(x_train, y_train)
```

```python
# Accuracy for prediction on training data
training_data_prediction= model.predict(x_train)
print(training_data_prediction)
```

```python
# R square error
score_1 = metrics.r2_score(y_train, training_data_prediction )

# Mean Absolute error
score_2 = metrics.mean_absolute_error(y_train, training_data_prediction)

print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
```

Prediction on Test data

```python
# Accuracy on predicting the test data
test_data_prediction = model.predict(x_test)
# print(test_data_prediction)
```

```python
# R square error
score_1 = metrics.r2_score(y_test, test_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(y_test, test_data_prediction)

print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
```

Visualizing the Actual Prices and Predicted Prices

```python
plt.scatter(y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Predicted Price")
plt.show()
```
